# BLR Municipality Complaint Data Analysis

---

## Objective
The goal of this analysis is to explore civic complaint data from Bengaluru and identify patterns in complaint categories, resolution status, geographical distribution, and trends over time.

This analysis helps understand:
• Which civic issues are most frequently reported  
• Which departments handle the most complaints  
• Areas with high complaint density  
• Complaint resolution efficiency  
• Trends in civic issues over time  

---

## Dataset Overview

The dataset contains citizen-reported civic complaints including information about:

• Complaint category  
• Sub-category  
• Department handling the issue  
• Complaint status  
• Location details (ward, latitude, longitude)  
• Timestamp of complaint  
• Number of comments on complaint  

This dataset helps analyze urban civic problem patterns.

---

## Column Description

| Column | Description |

## Insight 1 — Most discussed complaints

Higher comment volume suggests either greater public concern, longer resolution times, or wider geographic impact. Authorities may consider prioritizing these areas to improve citizen satisfaction and service efficiency.

## Insight 6 — Fastest resolving category

```python
# calculate resolution rate per category

status_table = pd.crosstab(
    df['category_title'],
    df['complaint_status_title']
)

# resolution rate = resolved / total complaints
status_table['resolution_rate'] = (
    status_table['Resolved'] /
    status_table.sum(axis=1)
)

fastest_resolving = (
    status_table['resolution_rate']
    .sort_values(ascending=False)
    .head(10)
)

fastest_resolving
```

## Visualization for insight 6

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))

bars = plt.barh(
    fastest_resolving.index,
    fastest_resolving.values
)

for bar in bars:
    width = bar.get_width()
    plt.text(
        width,
        bar.get_y() + bar.get_height()/2,
        f"{width:.2f}",
        va='center'
    )

plt.title("Fastest Resolving Complaint Categories")

plt.xlabel("Resolution Rate")

plt.ylabel("Category")

plt.gca().invert_yaxis()

plt.tight_layout()

plt.show()
```

##  Trend Insight — Complaint backlog trend (Open vs Resolved over time)

```python
# create year column (already converted earlier)
df['year'] = df['created_at'].dt.year

# group complaints by year and status
trend = pd.crosstab(
    df['year'],
    df['complaint_status_title']
)

trend
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

# plot open complaints
plt.plot(
    trend.index,
    trend['Open'],
    marker='o',
    label='Open'
)

# plot resolved complaints
plt.plot(
    trend.index,
    trend['Resolved'],
    marker='o',
    label='Resolved'
)

# plot re-opened complaints
if 'Re-opened' in trend.columns:
    plt.plot(
        trend.index,
        trend['Re-opened'],
        marker='o',
        label='Re-opened'
    )

plt.title("Complaint Resolution Trend Over Time")

plt.xlabel("Year")

plt.ylabel("Number of Complaints")

plt.legend()

plt.tight_layout()

plt.show()
```

The trend shows how complaint resolution has changed over time. While the number of complaints increased initially, the number of resolved cases also increased, indicating improvements in service efficiency. However, the presence of open and re-opened complaints suggests scope for faster response and improved resolution mechanisms.
